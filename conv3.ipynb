```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.zeros(200)
a[75:125] = 1
```

```python
plt.plot(a)
```

```python
b = np.zeros(200)
b[75:125] = np.linspace(1, 0, 50)
plt.plot(b)
```

```python
c = np.convolve(a,b)
```

```python
plt.plot(c)
```

```python
d = np.ones(200)
plt.plot(np.convolve(d,c))
```

```python
#Properties
# Commutativity
rtol=1e-05; atol=1e-08
assert np.allclose(np.convolve(a,b), np.convolve(b,a), rtol=rtol, atol=atol)
# Associativity
assert np.allclose(np.convolve(a, np.convolve(b, d)), np.convolve(np.convolve(a, b), d), rtol=rtol, atol=atol)
# Distirbutivity
assert np.allclose(np.convolve(a,b + d), np.convolve(a,b) + np.convolve(a,d), rtol=rtol, atol=atol)
```
